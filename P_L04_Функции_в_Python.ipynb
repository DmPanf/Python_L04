```python
def simple_separator():
    """
    –§—É–Ω–∫—Ü–∏—è —Å–æ–∑–¥–∞–µ—Ç –∫—Ä–∞—Å–∏–≤—ã–π —Ä–µ–∑–¥–µ–ª–∏—Ç–µ–ª—å –∏–∑ 10-–∏ –∑–≤–µ–∑–¥–æ—á–µ–∫ (**********)
    :return: **********
    """
    return '*' * 10


print(simple_separator() == '**********')  # True
```

```python
def long_separator(count):
    """
    –§—É–Ω–∫—Ü–∏—è —Å–æ–∑–¥–∞–µ—Ç —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å –∏–∑ –∑–≤–µ–∑–¥–æ—á–µ–∫ —á–∏—Å–ª–æ –∫–æ—Ç–æ—Ä—ã—Ö –º–æ–∂–Ω–æ —Ä–µ–≥—É–ª–∏—Ä–æ–≤–∞—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–º count
    :param count: –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–≤–µ–∑–¥–æ—á–µ–∫
    :return: —Å—Ç—Ä–æ–∫–∞ —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å, –ø—Ä–∏–º–µ—Ä—ã –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –Ω–∏–∂–µ
    """
    return '*' * count


print(long_separator(3) == '***')  # True
print(long_separator(4) == '****')  # True
```

```python
def separator(simbol, count):
    """
    –§—É–Ω–∫—Ü–∏—è —Å–æ–∑–¥–∞–µ—Ç —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å –∏–∑ –ª—é–±—ã—Ö —Å–∏–º–≤–æ–ª–æ–≤ –ª—é–±–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞
    :param simbol: —Å–∏–º–≤–æ–ª —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—è
    :param count: –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–π
    :return: —Å—Ç—Ä–æ–∫–∞ —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å –ø—Ä–∏–º–µ—Ä—ã –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –Ω–∏–∂–µ
    """
    return simbol * count


print(separator('-', 10) == '----------')  # True
print(separator('#', 5) == '#####')  # True
```

```python
def hello_world():
    """
    –§—É–Ω–∫—Ü–∏—è –ø–µ—á–∞—Ç–∞–µ—Ç Hello World –≤ —Ñ–æ—Ä–º–∞—Ç–µ:
    **********
    Hello World!
    ##########
    :return: None
    """
    my_word = '''
    **********

    Hello World!
    
    ##########'''
    print(my_word)


'''
**********
Hello World!
##########
'''
hello_world()
```

```python
def hello_who(who='World'):
    """
    –§—É–Ω–∫—Ü–∏—è –ø–µ—á–∞—Ç–∞–µ—Ç –ø—Ä–∏–≤–µ—Ç—Å—Ç–≤–∏–µ –≤ –∫—Ä–∞—Å–∏–≤–æ–º —Ñ–æ—Ä–º–∞—Ç–µ
    **********
    Hello {who}!
    ##########
    :param who: –∫–æ–≥–æ –º—ã –ø—Ä–∏–≤–µ—Ç—Å—Ç–≤—É–µ–º, –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é World
    :return: None
    """
    my_word = f'''
    **********
    Hello {who}!
    ##########'''
    print(my_word)


'''
**********
Hello World!
##########
'''
hello_who()
'''
**********
Hello Max!
##########
'''
hello_who('Max')
'''
**********
Hello Kate!
##########
'''
hello_who('Kate')
```

```python
def pow_many(power, *args):
    """
    –§—É–Ω–∫—Ü–∏—è —Å–∫–ª–∞–¥—ã–≤–∞–µ—Ç –ª—é–±–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ü–∏—Ñ—Ä –∏ –≤–æ–∑–≤–æ–¥–∏—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ —Å—Ç–µ–ø–µ–Ω—å power (–ø—Ä–∏–º–µ—Ä—ã –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –Ω–∏–∂–µ)
    :param power: —Å—Ç–µ–ø–µ–Ω—å
    :param args: –ª—é–±–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ü–∏—Ñ—Ä
    :return: —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤—ã—á–∏—Å–ª–µ–Ω–∏—è # True -> (1 + 2)**1
    """
    return sum(args) ** power 



print(pow_many(1, 1, 2) == 3)  # True -> (1 + 2)**1 == 3
print(pow_many(1, 2, 3) == 5)  # True -> (2 + 3)**1 == 5
print(pow_many(2, 1, 1) == 4)  # True -> (1 + 1)**2 == 4
print(pow_many(3, 2) == 8)  # True -> 2**3 == 8
print(pow_many(2, 1, 2, 3, 4) == 100)  # True -> (1 + 2 + 3 + 4)**2 == 10**2 == 100
```

```python
def print_key_val(**kwargs):
    """
    –§—É–Ω–∫—Ü–∏—è –≤—ã–≤–æ–¥–∏—Ç –ø–µ—Ä–µ–¥–∞–Ω–Ω—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –≤ —Ñ–∏–¥–µ key --> value
    key - –∏–º—è –ø–∞—Ä–∞–º–µ—Ç—Ä–∞
    value - –∑–Ω–∞—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞
    :param kwargs: –ª—é–±–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–º–µ–Ω–æ–≤–∞–Ω–Ω—ã—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
    :return: None
    """
    for key, val in kwargs.items():
      print(f'{key} --> {val}')


"""
name --> Max
age --> 21
"""
print_key_val(name='Max', age=21)
"""
animal --> Cat
is_animal --> True
"""
print_key_val(animal='Cat', is_animal=True)
```

```python
def my_filter(iterable, function):
    """
    (–£—Å–ª–æ–∂–Ω–µ–Ω–æ–µ –∑–∞–¥–∞–Ω–∏–µ —Å–æ *)
    –§—É–Ω–∫—Ü–∏—è —Ñ–∏–ª—å—Ç—Ä—É–µ—Ç –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å iterable –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –Ω–æ–≤—É—é
    –ï—Å–ª–∏ function –æ—Ç —ç–ª–µ–º–µ–Ω—Ç–∞ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç True, —Ç–æ —ç–ª–µ–º–µ–Ω—Ç –≤—Ö–æ–¥–∏—Ç –≤ –Ω–æ–≤—É—é –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∏–Ω–∞—á–µ –Ω–µ—Ç
    (–ø—Ä–∏–º–µ—Ä—ã –Ω–∏–∂–µ)
    :param iterable: –≤—Ö–æ–¥–∞–Ω—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏
    :param function: —Ñ—É–Ω–∫—Ü–∏—è —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏–∏
    :return: –Ω–æ–≤–∞—è –æ—Ç—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω–∞—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å
    """
    return list(filter(function, iterable))


print(my_filter([1, 2, 3, 4, 5], lambda x: x > 3) == [4, 5])  # True
print(my_filter([1, 2, 3, 4, 5], lambda x: x == 2) == [2])  # True
print(my_filter([1, 2, 3, 4, 5], lambda x: x != 3) == [1, 2, 4, 5])  # True
print(my_filter(['a', 'b', 'c', 'd'], lambda x: x in 'abba') == ['a', 'b'])  # True
```

```python
# –ü—Ä–æ–≥—Ä–∞–º–º–∞ "–õ–∏—á–Ω—ã–π —Å—á–µ—Ç"
yes_no = True
user_bank = 0.
user_purchase = {}


def choice_menu():  # menu
    global yes_no
    my_menu = {'1': '–ø–æ–ø–æ–ª–Ω–µ–Ω–∏–µ —Å—á–µ—Ç–∞', '2': '–ø–æ–∫—É–ø–∫–∞', '3': '–∏—Å—Ç–æ—Ä–∏—è –ø–æ–∫—É–ø–æ–∫', '4': '–±–∞–ª–∞–Ω—Å', '5': '—Å—Ä–µ–¥–Ω–∏–π —á–µ–∫', '0': '–≤—ã—Ö–æ–¥'}
    print('.' * 21)
    for key, val in my_menu.items():
        print(f'{key}. {val}')
    print('.' * 21)
    to_do = input('–í—ã–±–µ—Ä–∏—Ç–µ –ø—É–Ω–∫—Ç –º–µ–Ω—é: ')
    return yes_no, to_do


def bank_account():  # 1. –ø–æ–ø–æ–ª–Ω–µ–Ω–∏–µ —Å—á–µ—Ç–∞
    global user_bank
    try:
        new_sum = float(input('–í–≤–µ–¥–∏—Ç–µ —Å—É–º–º—É –¥–ª—è –ø–æ–ø–æ–ª–Ω–µ–Ω–∏—è —Å—á–µ—Ç–∞: '))
        if new_sum >= 10:
            user_bank += new_sum
        else:
            print('–°—É–º–º–∞ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –Ω–µ –º–µ–Ω–µ–µ 10 ..\n')
    except:
        print('–û—à–∏–±–∫–∞ –≤–≤–æ–¥–∞ —Å—É–º–º—ã!')


def make_purchase():  # 2. –ø–æ–∫—É–ø–∫–∞
    global user_bank
    global user_purchase
    try:
        purchase_sum = float(input('–í–≤–≤–µ–¥–∏—Ç–µ —Å—É–º–º—É –ø–æ–∫—É–ø–∫–∏: '))
        if purchase_sum < 10:
            print('–°—Ç–æ–∏–º–æ—Å—Ç—å –ø–æ–∫—É–ø–∫–∏ –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –º–µ–Ω–µ–µ 10 ..')
        elif purchase_sum <= user_bank:
            user_bank -= purchase_sum
            purchase = input('–í–≤–µ–¥–∏—Ç–µ –Ω–∞–∑–≤–∞–Ω–∏–µ –ø–æ–∫—É–ø–∫–∏: ')
            user_purchase[purchase] = purchase_sum
        else:
            print('–î–µ–Ω–µ–≥ –Ω–µ —Ö–≤–∞—Ç–∞–µ—Ç!')
    except:
        print('–û—à–∏–±–∫–∞ –≤–≤–æ–¥–∞ —Å—É–º–º—ã!')


def user_histroy():  # 3. –∏—Å—Ç–æ—Ä–∏—è –ø–æ–∫—É–ø–æ–∫
    if len(user_purchase) > 0:
        for key, val in user_purchase.items():
            print(f' - –°—Ç–æ–∏–º–æ—Å—Ç—å {key} = {val}')
        print('–°–¥–µ–ª–∞–Ω–æ –ø–æ–∫—É–ø–æ–∫: ', len(user_purchase))
        average_check()
        bank_balance()
        print()
    else:
        print('–ï—â–µ –Ω–µ –±—ã–ª–æ –ø–æ–∫—É–ø–æ–∫!')


def bank_balance():  # 4. –±–∞–ª–∞–Ω—Å
    print('üí∞ –û—Å—Ç–∞—Ç–æ–∫ –Ω–∞ —Å—á–µ—Ç—É:', user_bank)


def average_check():  # 5. —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫
    if len(user_purchase) > 0:
        print('üõí –°—Ä–µ–¥–Ω–∏–π —á–µ–∫ = ', round(sum(user_purchase.values()) / len(user_purchase), 1))
    else:
        print('–ï—â–µ –Ω–µ –±—ã–ª–æ –ø–æ–∫—É–ø–æ–∫!')


while yes_no:
    yes_no, choice = choice_menu()
    if choice == '1':
        bank_account()
    elif choice == '2':
        make_purchase()
    elif choice == '3':
        user_histroy()
    elif choice == '4':
        bank_balance()
    elif choice == '5':
        average_check()
    elif choice == '0':
        yes_no = False
    else:
        print('–ù–µ–≤–µ—Ä–Ω—ã–π –ø—É–Ω–∫—Ç –º–µ–Ω—é\n')
```
